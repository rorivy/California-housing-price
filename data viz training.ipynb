```python
cols = ['survived', 'sex', 'pclass', 'sibsp', 'parch', 'embarked', 'deck', 'embark_town', 'alone']
```

```python
sns.countplot(df_titanic['survived'])

plt.ylabel('count')
plt.title('Count plot of survived')
```

```python
for features in cols:
    sns.countplot(df_titanic[features], hue=df_titanic['survived'])
    plt.show()
```

```python
nr_rows = 3
nr_cols = 3

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(15, 15))

for r in range(0, nr_rows):
    for c in range(0, nr_cols):
        
        i = r * nr_cols + c
        ax = axs[r][c]
        
        sns.countplot(df_titanic[cols[i]], hue=df_titanic["survived"], ax=ax)
        ax.set_title(cols[i], fontsize=12, fontweight='bold')
        ax.legend(title="survived", loc= 'upper center')
```
